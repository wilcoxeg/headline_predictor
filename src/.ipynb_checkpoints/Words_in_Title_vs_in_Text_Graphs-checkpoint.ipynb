```python
import json
import time
import re
```

```python
data_filename = '../data/counted/raw_april.json'
start = time.clock() 
with open(data_filename) as in_f:
    data = json.load(in_f)
    
end = time.clock()
elapsed = end - start
print("It took to load data: %s seconds"% str(elapsed))
```

```python
start = time.clock()

do_words_appear = []
total_succ = 0
total_fail = 0
missed = []
succeeded = []
publisher_global_count = {}
publisher_succ_count = {}
for i,article in enumerate(data):
    h = article["headline"].lower()
    h_without_punc = re.sub(r'[^a-zA-Z0-9 ]','', h)
    text = article["plaintext"].lower()
    publisher = article["publisher"]
    url = article["url"]
    publisher_global_count[publisher] = 1 + publisher_global_count.get(publisher,0)
    
    headline_words = set(h_without_punc.split())
    text_words = set(text.split())
    intr = headline_words & text_words
    num_succ = len(intr)
    num_fail = len(headline_words) - num_succ
    if num_fail > 0:
        missed_parts = headline_words - intr
        missed += [(i,missed_parts, h, publisher)]
    else:
        succeeded += [(i,h,publisher)]
        publisher_succ_count[publisher] = 1 + publisher_succ_count.get(publisher,0)

    do_words_appear += (num_succ, num_fail)
    total_succ += num_succ
    total_fail += num_fail
    
print ("It took to process data (counting overlap b/w article and body): %s seconds" % (time.clock() - start))
```

```python
publisher_acc = [(publisher, float(publisher_succ_count.get(publisher,0)) / publisher_global_count[publisher], publisher_global_count[publisher]) for publisher in publisher_global_count.keys()]
publisher_acc = sorted(publisher_acc,key=lambda x: x[1], reverse=True)
publisher_acc[:10]
```

```python
acc_dict = {}
for p,a,c in publisher_acc:
    acc_dict[p] = a
```

```python
from matplotlib import pyplot as plt
import pylab

        

total_publisher_sum = reduce(lambda x,y: x+y, publisher_global_count.values())
publisher_perc = {}
for p, c in publisher_global_count.iteritems():
    publisher_perc[p] = float(c) / total_publisher_sum

biggest_data_all = sorted([(key, publisher_perc[key]) for key in publisher_perc], key=lambda x: x[1], reverse=True)
contrib_data = {}
for pub, contrib in biggest_data_all:
    contrib_data[pub] = contrib
biggest_data_chosen = biggest_data_all[:42]
biggest_data_chosen_top = biggest_data_all[:12]
figure(2, figsize=(9,9))
ax = axes([0.1, 0.1, 0.8, 0.8])

labels = [key[0] for key in biggest_data_chosen]
fracs = [key[1] for key in biggest_data_chosen]
remaining = 1 - sum(fracs)
labels += ["other"]
fracs += [remaining]
pie (fracs, labels=labels, autopct='%1.1f%%', startangle=90)
show()

top_producing_publishers = [key[0] for key in biggest_data_chosen_top]
top_producing_publisher_acc = {}

labels_acc = []
val = []
acc_tuples = []
for publisher, acc, count in publisher_acc:
    if publisher in top_producing_publishers:
        labels_acc += [publisher]
        val += [acc]
        acc_tuples += [(publisher, acc)]
    
fig = plt.figure()
ax = fig.add_subplot(111)
scatter_x = []
scatter_y = []
for pub in contrib_data:
    scatter_y += [contrib_data[pub]]
    scatter_x += [acc_dict[pub]]

ax.scatter(scatter_x, scatter_y,)
ax.set_xlabel('accuracy')
ax.set_ylabel('contribution')
plt.show()
```
